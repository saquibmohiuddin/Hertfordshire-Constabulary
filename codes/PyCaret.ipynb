```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/train_final-one-hot-mean-encoded.csv")
df
```

```python
from pycaret.classification import *
s = setup(data=df, target = 'outcome_type', 
          log_experiment = True, experiment_name = 'gold', verbose=True, use_gpu = False)
```

```python
best = compare_models(sort = "F1")
```

https://pycaret.gitbook.io/docs/get-started/functions/analyze

```python
best_tune = tune_model(best, optimize = "F1", choose_better = True, 
                       search_library = 'optuna', search_algorithm = "random")
```
